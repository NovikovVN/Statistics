```python
import pandas as pd
```

## Загрузка и обработка исходных данных

```python
mojo1 = pd.read_excel('mojo1.xlsx', index_col='Rank',
                       names=['Title', 'Studio', 'Gross / Box', 'Gross / Theaters',
                              'Opening / Box', 'Opening / Theaters', 'Release'])
mojo1
```

```python
def box_to_float(series):
    return series.apply(lambda s: float(s[1:].replace(',', '')))

def theaters_to_float(series):
    return series.apply(lambda s: int(str(s).replace('.', '')))
```

```python
mojo1.Release = pd.to_datetime(mojo1.Release)
mojo1['Gross / Box'] = box_to_float(mojo1['Gross / Box'])
mojo1['Gross / Theaters'] = theaters_to_float(mojo1['Gross / Theaters'][mojo1['Gross / Theaters'] != '-'])
mojo1['Opening / Box'] = box_to_float(mojo1['Opening / Box'][mojo1['Opening / Box'].notnull()])
mojo1['Opening / Theaters'] = theaters_to_float(mojo1['Opening / Theaters'][mojo1['Opening / Theaters'] != '-'])
mojo1 = mojo1.fillna(0.)
mojo1[['Gross / Theaters', 'Opening / Theaters']] = mojo1[['Gross / Theaters', 'Opening / Theaters']].astype(int)

mojo1
```

```python
mojo2 = pd.read_excel('mojo2.xlsx', index_col='Rank',
                      names=['Title', 'Studio', 'Adjusted Gross', 'Unadjusted Gross', 'Release'])

mojo2
```

```python
mojo2.Release = pd.to_datetime(mojo2.Release)
mojo2['Adjusted Gross'] = box_to_float(mojo2['Adjusted Gross'])
mojo2['Unadjusted Gross'] = box_to_float(mojo2['Unadjusted Gross'])

mojo2
```

```python
df = pd.merge(mojo1, mojo2[['Title', 'Adjusted Gross']], how='inner', on='Title')

df
```

```python
imdb = pd.read_excel('imdb.xlsx')

imdb
```

## Создание итогового набора данных и запись его в файл

```python
df = pd.merge(df, imdb, how='inner', on='Title').fillna('-')
df.set_index('Release', inplace=True)
df.sort_index(inplace=True)

df
```

```python
df.to_csv('mojo_imdb.csv', encoding='utf-8')
```
